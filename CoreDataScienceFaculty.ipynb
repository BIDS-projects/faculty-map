```python
import numpy as np
from datascience import *
```

```python
# Loads the csv data file for further data cleaning
file_name = "cleaned_researchers.csv"
researchers = Table.read_table(file_name)
researchers.sort("Name", descending=False)
```

```python
def dept_finder(t, column_name):
    related = []
    for dept in t.column(column_name): 
        dept = str(dept)
        if "Math" in dept or "Computer Science" in dept or "Statistics" in dept or "Information" in dept or "Operations Research" in dept:    
            related.append(True)
        else: 
            related.append(False)
    return t.with_column("Relevant Dept", related)
related_faculty = dept_finder(researchers, "Department").where("Relevant Dept", True)


data_science_key_terms = ["computation", "statistical analysis", "statistical computing", "statistical machine learning", "analytic", "database", "visualization", "data science", "probability", "data analysis", "data mining", "inference", "information management", "information retrieval", "information visualization", "information filtering", "modeling"]
is_data_related = []

def is_in_terms(interest, terms):
    for term in terms:
        if term in interest:
            return True
    return False

for interest in related_faculty.column("Interests"): 
    is_data_related.append(is_in_terms(interest, data_science_key_terms))

data_related_faculty = related_faculty.with_column("Data Related?", is_data_related).where("Data Related?", True).drop("Relevant Dept").drop("Data Related?")
data_related_faculty.show()
```

```python
data_related_faculty.to_df().to_csv('Core Data Science Faculty.csv', index = False)
```
